# Hacker News Project

In this project we will compare two types of posts `Ask HN` and `Show HN` to determine the following:

* Do Ask HN or Show HN receive more comments on average?
* Do posts created at a certain time receive more comments on average?


### Dataset description

The orginal dataset can be found [here](https://www.kaggle.com/hacker-news/hacker-news-posts).

It has been reduced to approximately 20,000 rows by removing all submissions that did not receive any comments, and then randomly sampling from the remaining submissions

Below are descriptions of the columns:

|Col no|Column Name| Column Description|
|-----|-----|:-----:|
|0| id:| The unique identifier from Hacker News for the post|
|1| title:| The title of the post|
|2| url:| The URL that the posts links to, if it the post has a URL|
|3| num_points: |The number of points the post acquired, calculated as the total number of upvotes minus the total number of downvotes|
|4| num_comments:| The number of comments that were made on the post|
|5| author:| The username of the person who submitted the post|
|6| created_at: | The date and time at which the post was submitted| 

# Data analysis

We start by converting the csv file into a list of lists.

```python
from csv import reader
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)

#print example five rows of the file
print(hn[0:5])
```

```python
#separate the header
headers = hn[0]
print(headers)
```

```python
#separate the rest of the datset
hn = hn[1:]
print(hn[:5])
```

## Filtering
We need to find posts with titles begining with `Ask HN` or `Show HN`. Well create a set of new lists containing just the data for those titles.

```python
#create empty lists
ask_posts =[]
show_posts = []
other_posts = []

for row in hn:
    #pick the column wher the title is
    title = row[1]
    #make it lowercase
    title = title.lower()
    if title.startswith('ask hn'):
        ask_posts.append(row)
    if title.startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
        
print('The number of ask posts is', len(ask_posts))
print('The number of show posts is', len(show_posts))
print('The number of other posts is', len(other_posts))
```

```python
#investigate 5 first lines of ask posts
print(ask_posts[:5])
```

```python
#investigate 5 first lines of show posts
print(show_posts[:5])
```

## Calculating comments
Now we'll need to find the averane number of comments received by each of the types of posts. 

```python
#loop over the list of ask posts and
#calculate the total number of comments
total_ask_comments = 0

for row in ask_posts:
    #find where the number is & turn into integer
    num_comments = int(row[4])
    #increment the total
    total_ask_comments += num_comments

print(total_ask_comments)

#clculate average nubmber of comments per post
avg_ask_comments=round(total_ask_comments / len(ask_posts))

print('The average number of comments per ask post is', avg_ask_comments) 
      
```

```python
#loop over the list of show posts and
#calculate the total number of comments
total_show_comments = 0

for row in show_posts:
    num_comments = int(row[4])
    total_show_comments += num_comments

    
print("The total number of comments received by 'show' posts is", total_show_comments)

#calculate averag
avg_show_comments=round(total_show_comments / len(show_posts))

print('The average number of comments per show post is', avg_show_comments) 
print('The diffrence between the average number of comments for "ask" and "show" posts is', avg_ask_comments - avg_show_comments)
```

On average an `ask post` receives 4 more comments than a `show post`.

## Time of posting analysis

Since `ask posts` are more likely to receive comments, we'll focus our remaining analysis just on these posts.

We'll determine if `ask posts` created at a certain time are more likely to attract comments.

We'll use the following steps:

1. Calculate the amount of ask posts created in each hour of the day, along with the number of comments received.
2. Calculate the average number of comments ask posts receive by hour created.

### Step 1

The date and time of postingis stored in column 6. The data format in col 6 is the following:

```python
print(ask_posts[1][6])
```

```python
# input is the dates and times
# output is the number of posts created in each hour

import datetime as dt 

#list of all lists
result_list = []

for row in ask_posts:
    #list of date+time wth number of comments
    value_pair =[]
    #idenfity the col where the date is
    created_at = row[6]
    #identify the col with the number of comments & turn into an integer
    num_comments = int(row[4])
    value_pair.append(created_at)
    value_pair.append(num_comments)
    result_list.append(value_pair)
    
print(list(result_list))
```

```python
# create two dictionaries  
#1 how many posts we have each hour
counts_by_hour = {}
#2 how many comments we get per each hour
comments_by_hour = {}
    
#iterate by the list of lists result_list    
for row in result_list:
    date_time = row[0]
    num_comments = row[1]
#convert the date time in to a datetime object
    date_dt = dt.datetime.strptime(date_time, "%m/%d/%Y %H:%M")
#select the hour    
    hour_dt = dt.datetime.strftime(date_dt,"%H")
    if hour_dt not in counts_by_hour:
#create a key in the dictionary; it would output sth like 09 : 10 
#that is e.g. there are 10 posts made between 9-9.59am
        counts_by_hour[hour_dt] = 1
#key in the 2nd dictionary, hour is the key, value is the number of comments    
        comments_by_hour[hour_dt] = num_comments
    else:
        counts_by_hour[hour_dt] += 1
        comments_by_hour[hour_dt] += num_comments

#dictionary contains the number of ask posts created during each hour of the day        
print(counts_by_hour)
```

```python
#dictionary contains the hour and the corresponding number of comments the posts (posted at this hour) have received
print(comments_by_hour)
```

### Step 2
Calculate the average number of comments for posts created during each hour of the day.

```python
#create a list of lists with avg values of comments per post per hour
avg_by_hour = []

for hour in comments_by_hour:
    #divide all comments made each hour by the number of posts created during this hour
    avg_by_hour.append([hour, comments_by_hour[hour]/counts_by_hour[hour]])

print(avg_by_hour)
```

Sort the list of lists and print the five highest values in a format that's easier to read.

```python
#create a list of lists with swapped values hour:average > average:hour
swap_avg_by_hour = []

for row in avg_by_hour:
    #create a new list
    value_pair = []
    average = row[1]
    hour = row[0]
    #first append the average, then the hour
    value_pair.append(average)
    value_pair.append(hour)
    #append the pair to the master list
    swap_avg_by_hour.append(value_pair)
    
print(swap_avg_by_hour)
```

```python
#sort the list
sorted_swap= sorted(swap_avg_by_hour, reverse= True)

print("Top 5 Hours for Ask Posts Comments:")

for row in sorted_swap[:5]:
    average = row[0]
    hour = row[1]
    #format the number into an hour and display as hour:min
    dt_hour = dt.datetime.strptime(hour, "%H")
    hour = dt.datetime.strftime(dt_hour, "%H:%M")
    print("{}: {:.2f} average comments per post".format(hour,average))
```
